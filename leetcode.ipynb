```python
from collections import deque

class BST:
    """
    Binary search tree with as many functions as I can think of as fast 
    Includes an instance variable called 'order' that stores the level 
    of a node
    """
    def __init__(self, data, left=None, right=None, order=0):
        self.data = data
        self.left = left
        self.right = right
        self.order = order

    def add_child(self, data):

        """
        Don't accept duplicates
        """
        if data == self.data:
            return
        if data < self.data:
            if self.left:
                self.left.add_child(data)
            else:
                self.left = BST(data, order=self.order + 1)
        elif data > self.data:
            if (self.right):
                self.right.add_child(data)
            else:
                self.right = BST(data, order=self.order + 1)
    
    def search(self, val):
        """
        search a BST for a value 'val'. Return True if found, False if not found
        """
        if val == self.data:
            return True
            
        if val < self.data:
            if self.left:
                return self.left.search(val)
            else:
                return False
                
        elif (val > self.data):
            if self.right:
                return self.right.search(val)
            else:
                return False

    def IOT(self):
        """
        In-order traversal - return the elements in a list
        """
        elements = []
        
        if self.left:
            elements += self.left.IOT()

        elements.append(self.data)

        if self.right:
            elements += self.right.IOT()

        return elements

    def PrOT(self):
        """
        Pre-order traversal - return the elements in a list
        """
        elements = []
        elements.append(self.data)
        if self.left:
            elements += self.left.IOT()
        if self.right:
            elements += self.right.IOT()

        return elements
            
    def PoOT(self):
        """
        Post-order traversal, return the ele,ents in a list
        """
        elements = []
        if self.left:
            elements += self.left.IOT()
        if self.right:
            elements += self.right.IOT()
        elements.append(self.data)
        return elements

    def LOT(self):
        """
        Level order traversal
        return the level with the data in a list
        """
        q = deque()
        elements = []
        q.append(self)

        while q:
            node = q.popleft()
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
            elements.append([node.data, node.order])
        return elements
        
    def find_min(self):
        """
        Return the minimum element
        """
        if self.left:
            return self.left.find_min()
    
        return self.data

    def find_max(self):
        """
        Return the maximum element
        """
        if self.right:
            return self.right.find_max()
        return self.data

    def delete(self, val):
        """
        delete the node with the value 'val'
        """
        if val < self.data:
            if self.left:
                self.left = self.left.delete(val)
        elif val > self.data:
            if self.right:
                self.right = self.right.delete(val)

        else:
            if (self.left == None) and (self.right == None):
                return None
            if self.left == None:
                return self.right
            if self.right == None:
                return self.left
            else:
                min_val = self.right.find_min()
                self.data = min_val
                self.right.delete(min_val)
                
        return self
            
    def mirror(self):
        """
        Mirror transform the BST on an axis doen the middle
        if there is one root , one left child and one right child, 
        then mirror will swap the left and the right child
        
        """
        if self.left and self.right:
            nodeleft = self.left
            self.left = self.right
            self.right = nodeleft

            self.left.mirror()
            self.right.mirror()

        elif self.right:
            self.left = self.right
            self.right = None
        elif self.left:
            self.right = self.left
            self.left = None

def build_tree(elements):
    print ("Building tree with these elements:", elements)
    root = BST(elements[0])
    for data in elements[1:]:
        root.add_child(data)

    return root
            
if __name__ == '__main__':
    celements = ['India', 'Pakistan', 'Germany', 'USA', 'China', 'India', 'UK', 'USA']
    ielements = [17, 4, 1, 20, 9, 23, 18, 34, 11, 21, 4, 8, 23]

    c_tree = build_tree(celements)
    print ("UK is in the list?" ,c_tree.search("UK"))
    print ("Sweden is in the list?", c_tree.search("Sweden"))

    i_tree = build_tree(ielements)
    print ("8 is in the list:", i_tree.search(8))
    print ("32 is in the list:", i_tree.search(32))
    print (i_tree.LOT())
    i_tree.delete(4)
    print (i_tree.LOT())
    i_tree.add_child(4)
    print (i_tree.LOT())
    i_tree.mirror()
    print(i_tree.LOT())
    i_tree.mirror()
    print(i_tree.LOT())
    i_tree.delete(20)
    print(i_tree.LOT())
    i_tree.delete(9)
    print(i_tree.LOT())
```

```python
560. Subarray Sum Equals K
Medium

Topics
Companies

Hint
Given an array of integers nums and an integer k, return the total number of subarrays whose sum equals to k.

A subarray is a contiguous non-empty sequence of elements within an array.

 

Example 1:

Input: nums = [1,1,1], k = 2
Output: 2
Example 2:

Input: nums = [1,2,3], k = 3
Output: 2


This problem can be solved by brute force method by generting all possible subarrays and calculating their sum. It would use two nested pointers, each looping over the indicesof the array, one holding the left index of a ubarray and the other holding the light index of a subarray. Indide the double loop a sum calculates the sum. This is O(n2)*O(n) = O(n3)

We can instead of doing a sum inside the two nested loops use a scalar variable initialized withing the outer array and add the next element of the array to it to replicate the sum. That mkes it O(n). 

The O(n) solution exists. 
Note that any possible subarray is defined by a beginning element and an end element. 

Let sum_{ij} be the sum of the subarray beginning at the ith index and ending at the jth index. 

We can calulate sum_{0j}  for all j (all subarrays beginning at the first element and ending at the j-th element) with one linear pass over the array. sum_{00} is the first element, and sum_{0j) = sum_{0,j-1} + array[j]. sum{0,i} is called the prefixSum of i. 

Now note that sum_{i,j} = sum{j,0} - sum{i,0}. 

In our problem, we need to find the condition sum{j,0} - sum{i,0} = k. j>i (to be a valid subarray). 
Or, sum_{j,0} - k = {sum{i,0}. 

We build the prefixSum of each array element and store them in a hash table where the prefixSum is key and its frequency is the value. Then we scan the array and calculate prefixSum for each array and check if prefixSum[i] + k is in the hash table. If it does we add out counter by the corresponding frequency. 

We can do slightly better:

Checking if prefixSum[i] + k exists in the hash table of prefixSum requires the whole hash table to be first build. 
But checking if prefixSum[i] - k exists in the hash table requires one pass. We can build the hash table as we go,  since

prefixSum[i] - k = prefixSum[j] implies i>j, so if there is indeed a matching prefixSum[j] it must already exist in the hash table. 

One caveat is that it could be that prefixSum[i] = k for one or more i. That is the subarray beginnign at the first element and ending at the ith element sums to j. That means the matching prefixSum[j] = 0. However, we havent added a 0 key to out hash table. 

We must first initialize the hash table with {0:1} or inser a zero in the beginning of the array (difficult to do with an array). 

Theese two methods are both O(n) time and O(n) space but one method uses a sinple loop and the other uses two consecute loops. the codes are below.
```

```python
### BST Range search

#Build a Binary search tree of integers. Take two numbers as inputs. Print all the numbers in the BST that 
#falls between those two numbers

from collections import deque

class bst:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

    def add_child(self, data):
        if data == self.data:
            return
        if data < self.data:
            if self.left:
                self.left.add_child(data)
            else:
                self.left = bst(data)
        if data > self.data:
            if self.right:
                self.right.add_child(data)
            else:
                self.right = bst(data)

    def IOT(self):
        elements = []
        if self.left:
            elements += self.left.IOT()

        elements.append(self.data)

        if self.right:
            elements +=self.right.IOT()

        return elements

    def PrOT(self):
            elements = []
            elements.append(self.data)
            if self.left:
                elements += self.left.IOT()

            if self.right:
                elements +=self.right.IOT()

            return elements

    def PoOT(self):
            elements = []
            if self.left:
                elements += self.left.IOT()

            if self.right:
                elements +=self.right.IOT()

            elements.append(self.data)

            return elements
    def LOT(self):
        q = deque()
        q.append(self)
        elements = []

        while q:
            node = q.popleft()
            elements.append(node.data)
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
        return elements
    def rangeT(self, a, b):
        elements = []
    
        if (self.data >=a) and (self.data <=b):
            if self.left:
                elements += self.left.rangeT(a,b)
                
            elements.append(self.data)
            #print (elements)           
            if self.right:
                elements += self.right.rangeT(a,b)
                
        elif (self.data < a):
                 
            if self.right:
                elements += self.right.rangeT(a,b)

        elif (self.data > b):
            if self.left:
                elements += self.left.rangeT(a,b)

        return elements    

def build_tree(elements):
    root = bst(elements[0])

    for i in elements[1:]:
        root.add_child(i)
    return root
        

elements = [87,1,2,9,8,3,0,3,7,65,2,99,5, 20, 21, 44, 33, 18, 27]

root = build_tree(elements)
print (root.IOT())
print (root.LOT())
print(root.rangeT(8,65))

```

```python
class sparseVec:
    def __init__(self, vec):
        self.vec = vec
        self.dict = {}
        for idx, element in enumerate(self.vec):
            if element:
                self.dict[idx] = element
        
    def sparsedot(self, sv1, sv2):
        prod = 0
        for index in sv1:
            if index in sv2:
                prod = prod + sv1[index]*sv2[index]
        return prod


A = [0,0,2,0,0,0,1,0,0,0,3]
B = [2,0,0,1,0,0,4,0,1,0,2]

A = sparseVec(A)
B = sparseVec(B)

sA = A.dict
sB = B.dict

A.sparsedot(sA, sB)
```
